*  타겟 설정 ( 노선 수 )
*  승차 승객 수 와 하차 승객 수를 합쳐 승하차 총 승객수를 만들어줌

```python
# 시각화 한글폰트 설정
import seaborn as sns
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
sns.set(font="Malgun Gothic",#"NanumGothicCoding", 
        rc={"axes.unicode_minus":False}, # 마이너스 부호 깨짐 현상 해결
        style='darkgrid')     
```

```python
# 원래는 상관계수 만이 절대적 지표는 아니지만 다른 신뢰성을 평가할만한 방법 x
corr = df.corr()
corr
```

* 값들 중 노선수 |corr| 값이 0.5 이상이 되는 것들을 선별

```python
corr2 = corr.loc[abs(corr.노선수) >= 0.5]
corr2[['노선수']]
```

* target값과 승차총승객수, 하차총 승객수, 평균이동시간이 관련있다고 판단
* 두 변수의 p-value 값도 비교해서 유의미성 판단

```python
import scipy.stats as spst
```

```python
print(spst.pearsonr(df.정류장수, df.승차총승객수))
print(spst.pearsonr(df.정류장수, df.하차총승객수))
print('-'*50)
print(spst.pearsonr(df.노선수, df.승차총승객수))
print(spst.pearsonr(df.노선수, df.하차총승객수))
```

* p-value값도 모두 < 0.05 (귀무가설 기각, 대립가설 채택)
* 이로서 정류장수, 노선수 는 승차총승객수, 하차총승객수와 관련이 있다 채택
* target 값과 변수를 나누어 비율 산정

```python
df['비율1'] = df['정류장수'] / df['승차총승객수'] * 10000
df.sort_values(by=['비율1']).head()
```

```python
df['비율2'] = df['정류장수'] / df['하차총승객수'] * 10000
df.sort_values(by=['비율1']).head()
```

```python
df['비율3'] = df['노선수'] / df['승차총승객수'] * 10000
df.sort_values(by=['비율3']).head()
```

```python
df['비율4'] = df['노선수'] / df['하차총승객수'] * 10000
df.sort_values(by=['비율4']).head()
```

* 비율 1 에서 '중구', '동대문구', '관악구', '강남', '종로' 가 1 ,2 ,3, 4, 5등
* 비율 2 에서 '중구', '관악구', '동대문구', '강남', '종로' 가 1, 2, 3, 4, 5등
* 비율 3 에서 '강동구', '송파구', '강서구', '관악구', '강남구' 가 1, 2, 3, 4, 5등
* 비율 4 에서 '강동구', '송파구', '강서구', '관악구', '강남구' 가 1, 2, 3, 4, 5등

- 따라서 비율 1등 중구, 강동구 / 비율4개가 모두 겹치는 관악구 선택
